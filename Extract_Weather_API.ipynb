```python
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize
from pandas import Timestamp

import pymongo
from pymongo import MongoClient
#from pandas import Timestamp

conn = 'mongodb://127.0.0.1:27017' # Initialize PyMongo to work with MongoDBs
client = pymongo.MongoClient(conn) # Set Client connection to server (daemon)

db = client.dockless_vehicle_db          # Select MongoDB database to work with
collection = db.weather_collection 
```

```python
url = 'http://hrly.lsu.edu/GetHrlyData/'
body = {
  "order": "asc",
  "offset": 0,
  "limit": 50,
  "sdate": "2018040100",
  "edate": "2019020200",
  "stime": "",
  "etime": "",
  "stn": "KAUS",
  "src": "",
  "interval": "1h",
  "elems": "temp,dewp,relative_humidity,percip_1h,wind_speed,press_sea_level",
  "summary": "temp__mean",
  "reduce": "temp__last",
  "where": ""
}
```

```python
json.dumps(body)
```

```python
#help(requests.post)
```

```python
response = requests.post(url, data=json.dumps(body))
```
